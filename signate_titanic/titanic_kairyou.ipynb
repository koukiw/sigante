```python
import pandas as pd
train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/signate_titanic/train.tsv", sep="\t", index_col=0) # 学習用データ
test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/signate_titanic/test.tsv", sep="\t", index_col=0)   # 評価用データ
sample_submit = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/signate_titanic/sample_submit.tsv", sep="\t", index_col=0, header=None) # 応募用サンプルファイル
```

```python
data = pd.concat([train, test])
```

```python
data
```

```python
data['age'] = data['age'].fillna(data['age'].mean()) 
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()) 
```

```python
data = pd.get_dummies(data)
data
```

```python
train = data.loc[train.index]
test = data.loc[test.index]
test = test.drop(["survived"], axis=1)
```

```python
train.head()
```

```python
y = train["survived"] # 目的変数
X = train.drop(["survived"], axis=1) # 目的変数を除いたデータ
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
```

```python
pred = model.predict_proba(test)[:, 1] 
pred.shape
```

```python
model.predict_proba(test)
```

```python
sample_submit[1] = pred
sample_submit.to_csv('submit.tsv', header=None, sep='\t')
```
